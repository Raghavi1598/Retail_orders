```python
# 1. find top highest revenue generating products

mycursor.execute("""SELECT product_id, sum(sale_price) as sales
FROM df_orders
GROUP BY product_id
ORDER BY sales DESC
LIMIT 10;""")

myresult_1 = mycursor.fetchall()
myresult_1    
```

```python
# 2) find the top 5 cities with the highest profit and margins
mycursor.execute("""SELECT city, SUM(profit) as total_profit 
FROM df_order 
JOIN df_orders ON df_order.order_id = df_orders.order_id 
GROUP BY city 
ORDER BY total_profit DESC 
LIMIT 5;""")
myresult_2= mycursor.fetchall()
myresult_2 
```

```python
#3) calculate the total discount given for each category
mycursor.execute("""SELECT category, SUM(discount) AS total_discount 
FROM df_order 
JOIN df_orders ON df_order.order_id = df_orders.order_id   
GROUP BY category 
LIMIT 0, 1000; """)
myresult_3= mycursor.fetchall()
myresult_3 
```

```python
#4) find the average sale price per product category
mycursor.execute("""SELECT category, avg(sale_price) AS avg_sale_price 
FROM df_order 
JOIN df_orders ON df_order.order_id = df_orders.order_id   
GROUP BY category 
LIMIT 0, 1000;""")
myresult_4= mycursor.fetchall()
myresult_4
```

```python
#5) find the region with the highest average sale price

mycursor.execute("""SELECT region, AVG(sale_price) AS avg_sale_price
FROM  df_order 
JOIN df_orders ON df_order.order_id = df_orders.order_id
GROUP BY region
ORDER BY avg_sale_price DESC
LIMIT 1;""")
myresult_5= mycursor.fetchall()
myresult_5

```

```python
#6) find the total profit per total category
mycursor.execute("""SELECT category, SUM(profit) AS total_profit 
FROM df_order 
JOIN df_orders ON df_order.order_id = df_orders.order_id   
GROUP BY category 
LIMIT 0, 1000; """)
myresult_6= mycursor.fetchall()
myresult_6 
```

```python
#7) identity the top 3 segments wuth the highest quantity of orders

mycursor.execute("""SELECT segment, SUM(quantity) AS total_quantity
FROM df_order 
JOIN df_orders ON df_order.order_id = df_orders.order_id    
GROUP BY segment
ORDER BY total_quantity DESC
LIMIT 3;""")
myresult_7= mycursor.fetchall()
myresult_7
```

```python
#8) determine the average discount percentage given per region
mycursor.execute("""SELECT region, AVG(discount) AS average_discount_percentage
FROM df_order 
JOIN df_orders ON df_order.order_id = df_orders.order_id
GROUP BY region;""")
myresult_8= mycursor.fetchall()
myresult_8
```

```python
# 9) find the product category with the highest total profit

mycursor.execute("""SELECT Category, SUM(Profit) AS TotalProfit
FROM df_order 
JOIN df_orders ON df_order.order_id = df_orders.order_id  
GROUP BY Category
ORDER BY TotalProfit DESC
LIMIT 1;""")
myresult_9= mycursor.fetchall()
myresult_9
```

```python
#10) calculate the total revenue generated per year

mycursor.execute("""SELECT YEAR(order_date) AS year, SUM(quantity) AS total_quantity
FROM df_order 
JOIN df_orders ON df_order.order_id = df_orders.order_id 
GROUP BY YEAR(order_date)
ORDER BY year;""")
myresult_10= mycursor.fetchall()
myresult_10
```

```python
# 3  Identify the minimum, maximum, and average values in each category
mycursor.execute("""select category,min(profit),max(profit),avg(profit) from df_order 
JOIN df_orders ON df_order.order_id = df_orders.order_id group by category;""")
myresult_13= mycursor.fetchall()

myresult_13
```

```python
# 4  Identify up to 10 subcategories within each category.
mycursor.execute("""select category,sub_category 
from df_order JOIN df_orders ON df_order.order_id = df_orders.order_id
limit 10;""")
myresult_14= mycursor.fetchall()
myresult_14
```

```python
#5 Combine two tables using a left join based on state and subcategory
mycursor.execute("""SELECT 
    df_order.order_id AS order_id, 
    df_order.state AS state,
    df_orders.sub_category AS sub_category
FROM 
    df_order
LEFT JOIN 
    df_orders 
ON 
    df_order.order_id = df_orders.order_id
LIMIT 10;""")
myresult_15= mycursor.fetchall()
myresult_15
```

```python
#6  Combine two tables using a right join based on state and subcategory
mycursor.execute("""SELECT 
    df_order.order_id AS order_id, 
    df_orders.sub_category AS sub_category,
    df_order.state AS state
FROM 
    df_order
RIGHT JOIN 
    df_orders 
ON 
    df_order.order_id = df_orders.order_id
LIMIT 10;""")
myresult_16= mycursor.fetchall()
myresult_16
```

```python
# 7)Identify the category where the subcategory is 'Art  (using sub query)
mycursor.execute("""select category from df_order where order_id
in (select order_id from df_orders where sub_category="Art") 
limit 3;""")
myresult_17= mycursor.fetchall()
myresult_17
```

```python
#8 Retrieve the top 10 regions and convert their names to uppercase.
mycursor.execute("""select upper(region) from df_order limit 10;""")
myresult_18= mycursor.fetchall()
myresult_18
```

```python
#9 Identify the top 50 ranks within the category(using windows fun)

mycursor.execute("""select sub_category, dense_rank() 
over (order by quantity desc) 
from df_orders
limit 50;""")
myresult_19= mycursor.fetchall()
myresult_19
```
